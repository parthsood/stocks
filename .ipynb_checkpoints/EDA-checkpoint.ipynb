```python
from nsepy import get_history
import datetime as dt
import pandas as pd
import matplotlib.pyplot as plt
```

```python
companies = pd.read_csv('list_of_symbols.csv')
```

```python
l = list(companies.Symbol)
print(l)
```

```python
t = get_history(symbol="SBIN", start=dt.date.today() - dt.timedelta(days=730), end=dt.date.today())
```

```python
plt.figure(figsize=(16,10))

plt.plot(t.index, t.Close, linewidth=2, label='Close')

plt.ylim(0,)
plt.legend()
plt.grid()
plt.show()
```
